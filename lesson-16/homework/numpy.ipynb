```python
def fahrenheit_to_celsius(F):
    return (F - 32) * 5 / 9
vectorized_conversion = np.vectorize(fahrenheit_to_celsius)
temps_f = np.array([32, 68, 100, 212, 77])
temps_c = vectorized_conversion(temps_f)

print("Temperatures in Fahrenheit:", temps_f)
print("Temperatures in Celsius:", temps_c)
```

```python
%pip install opencv-python
import cv2
import numpy as np

def brighten_channel(image_path, channel_index, brightness_increase, output_path):
    image = cv2.imread(image_path)
    if image is None:
        raise ValueError("Image not found or unable to read.")
    
    # Split the image into B, G, R channels
    channels = cv2.split(image)
    
    # Increase brightness in the specified channel
    channels[channel_index] = np.clip(channels[channel_index] + brightness_increase, 0, 255)
    
    # Merge the channels back
    brightened_image = cv2.merge(channels)
    
    # Save the modified image
    cv2.imwrite(output_path, brightened_image)
    print(f"Modified image saved as {output_path}")

# Example usage
brighten_channel('bird.jpg', 2, 40, 'masked_output.jpg')  # Increases red channel brightness by 40
```
