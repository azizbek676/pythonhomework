```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
a=[5,4,5,6,7]
b=[1,2,3,4,5]
plt.plot(a,b)  
```

1

```python
x = np.linspace(-10, 10, 400)
y = x**2 - 4*x + 4

plt.figure(figsize=(6, 4))
plt.plot(x, y, label=r'$f(x) = x^2 - 4x + 4$', color='b')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Basic Plot: Quadratic Function')
plt.legend()
plt.grid()
plt.show()
```

2

```python
x = np.linspace(0, 2 * np.pi, 400)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(6, 4))
plt.plot(x, y_sin, 'r--', label='sin(x)')
plt.plot(x, y_cos, 'b-', label='cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine and Cosine Plot')
plt.legend()
plt.grid()
plt.show()
```

3

```python
x = np.linspace(-2, 2, 100)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].plot(x, x**3, 'g')
axes[0, 0].set_title(r'$f(x) = x^3$')

axes[0, 1].plot(x, np.sin(x), 'b')
axes[0, 1].set_title(r'$f(x) = \sin(x)$')

plt.tight_layout()
plt.show()
```

4

```python
np.random.seed(0)
x_scatter = np.random.uniform(0, 10, 100)
y_scatter = np.random.uniform(0, 10, 100)

plt.figure(figsize=(6, 4))
plt.scatter(x_scatter, y_scatter, c='purple', marker='o', alpha=0.7)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Random Scatter Plot')
plt.grid()
plt.show()
```

5

```python
data = np.random.normal(0, 1, 1000)

plt.figure(figsize=(6, 4))
plt.hist(data, bins=30, color='blue', alpha=0.7, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Normally Distributed Data')
plt.grid()
plt.show()
```

6

```python
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.cos(X**2 + Y**2)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot: $cos(x^2 + y^2)$')

plt.show()
```

```python
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
sales = [200, 150, 250, 175, 225]
plt.figure(figsize=(6, 4))
plt.bar(products, sales, color=['red', 'blue', 'green', 'purple', 'orange'])
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Sales Data for Different Products')
plt.show()
```

```python
time_periods = ['T1', 'T2', 'T3', 'T4']
category_A = [5, 7, 8, 6]
category_B = [3, 5, 6, 4]
category_C = [2, 3, 4, 3]

fig, ax = plt.subplots(figsize=(6, 4))
ax.bar(time_periods, category_A, label='Category A', color='red')
ax.bar(time_periods, category_B, bottom=category_A, label='Category B', color='blue')
ax.bar(time_periods, category_C, bottom=np.array(category_A) + np.array(category_B), label='Category C', color='green')

ax.set_xlabel('Time Periods')
ax.set_ylabel('Values')
ax.set_title('Stacked Bar Chart')
ax.legend()

plt.show()
```
